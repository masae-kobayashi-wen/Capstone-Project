### 1.4.1 Missing Values By Columns

```python
# refer to '6_GuidedCapstone/02_data_wrangling_mk.ipynb'
# Count (using `.sum()`) the number of missing values (`.isnull()`) in each column 
# & calculate the percentages (using `.mean()` instead of `.sum()`)
#Order them (increasing or decreasing) using sort_values
#Call `pd.concat` to present these in a single table (DataFrame) with the helpful column names 'count' and '%'
# refer to 'https://datascience.stackexchange.com/questions/37878/difference-between-isna-and-isnull-in-pandas'
missing = pd.concat([spd.isnull().sum(), 100 * spd.isnull().mean()], axis=1) # isna() gives the same results
missing.columns=['count', '%']
missing.sort_values(by=['count', '%'])
```

```python
# columns with missing data
missing1 = missing[missing['count'] != 0]
missing1.sort_values(by=['count', '%'])
```

182 columns out of 195 columns have missing data

```python
# columns without missing data
missing_no = missing1 = missing[missing['count'] == 0]
missing_no
```
